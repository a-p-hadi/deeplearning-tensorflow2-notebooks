## <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">لود مجموعه داده (dataset)</div>

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
import numpy as np

from dataset import load_hoda
```

```python
x_train_original, y_train_original, x_test_original, y_test_original = load_hoda()
```

```python
# Preprocess input data for Keras. 
y_train = keras.utils.to_categorical(y_train_original, num_classes=10)
y_test = keras.utils.to_categorical(y_test_original, num_classes=10)
```

<div style="direction:rtl;text-align:right;font-family:Tahoma">
نرمال سازی مقادیر به مقدار بین 0 و 1 
</div>

```python
x_train = x_train_original.astype('float32')
x_test = x_test_original.astype('float32')
x_train /= 255
x_test /= 255
```

## <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">تعریف معماری مدل (model architecture)</div>

```python
model = Sequential()
model.add(Dense(64, activation='relu', input_dim=25))
model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```

## <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">Compile model</div>


```python
model.compile(loss='categorical_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])
```

## <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">آموش مدل با داده‌های آموزشی</div>


```python
history = model.fit(x_train, y_train,
          epochs=100,
          batch_size=64, validation_split=0.2)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(history.history['val_accuracy'])
```

## <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">ارزیابی مدل روی داده های آزمون</div>

```python
loss, acc = model.evaluate(x_test, y_test)
print('\nTesting loss: %.2f, acc: %.2f%%'%(loss, acc))
```

## <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">پیش‌بینی داده‌های آموزشی</div>


```python
# The predict_classes function outputs the highest probability class
# according to the trained classifier for each input example.
predicted_classes = model.predict(x_test)
predicted_labels = np.argmax(predicted_classes, axis=1)
print("predicted:")
print(predicted_labels)
print("True Label:")
print(y_test_original)
```

# <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">بدون تبدیل به one-hot</div>
<div style="direction:rtl;text-align:right;font-family:Tahoma">

```python
x_train_original, y_train_original, x_test_original, y_test_original = load_hoda()
y_train = y_train_original.astype('float32')
y_test = y_test_original.astype('float32')
x_train = x_train_original.astype('float32')
x_test = x_test_original.astype('float32')
x_train /= 255
x_test /= 255
model = Sequential()
model.add(Dense(64, activation='relu', input_dim=25))
model.add(Dense(10, activation='softmax'))
model.compile(loss='SparseCategoricalCrossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])
history = model.fit(x_train, y_train,
          epochs=100,
          batch_size=64, validation_split=0.2)
predicted_classes = model.predict(x_test)
predicted_labels = np.argmax(predicted_classes, axis=1)
```

```python
from sklearn.metrics import confusion_matrix
plt.matshow(confusion_matrix(y_test_original, predicted_labels), cmap="Blues")
plt.colorbar()
plt.xticks(range(10))
plt.yticks(range(10))
plt.xlabel("Predicted label")
plt.ylabel("True label");
```

```python
loss, acc = model.evaluate(x_test, y_test)
print('\nTesting loss: %.2f, acc: %.2f%%'%(loss, acc))
```

# <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">بدون Softmax</div>
<div style="direction:rtl;text-align:right;font-family:Tahoma">
<div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">قبل از اعمال softmax را logits می گوییم</div>
<div style="direction:rtl;text-align:right;font-family:Tahoma">

```python
import tensorflow as tf
x_train_original, y_train_original, x_test_original, y_test_original = load_hoda()
y_train = y_train_original.astype('float32')
y_test = y_test_original.astype('float32')
x_train = x_train_original.astype('float32')
x_test = x_test_original.astype('float32')
x_train /= 255
x_test /= 255
model = Sequential()
model.add(Dense(64, activation='relu', input_dim=25))
model.add(Dense(10))
model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              optimizer='rmsprop',
              metrics=['accuracy'])
history = model.fit(x_train, y_train,
          epochs=100,
          batch_size=64, validation_split=0.2)
predicted_classes = model.predict(x_test)
predicted_labels = np.argmax(predicted_classes, axis=1)
loss, acc = model.evaluate(x_test, y_test)
print('\nTesting loss: %.2f, acc: %.2f%%'%(loss, acc))
```

# <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">ورودی 2 بعدی به شبکه عصبی</div>
<div style="direction:rtl;text-align:right;font-family:Tahoma">

```python
from dataset import load_hoda_2d
from keras.layers import Flatten
x_train_original, y_train_original, x_test_original, y_test_original = load_hoda_2d()
x_train_original = np.array(x_train_original)
x_test_original = np.array(x_test_original)
y_train = y_train_original.astype('float32')
y_test = y_test_original.astype('float32')
x_train = x_train_original.astype('float32')
x_test = x_test_original.astype('float32')
x_train /= 255
x_test /= 255
model = Sequential()
model.add(Flatten(input_shape=(5,5))) # بدون تعیین شکل ورودی سامری نمیدهد
model.add(Dense(64, activation='relu'))
model.add(Dense(10))
model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              optimizer='rmsprop',
              metrics=['accuracy'])
history = model.fit(x_train, y_train,
          epochs=100,
          batch_size=64, validation_split=0.2)
predicted_classes = model.predict(x_test)
predicted_labels = np.argmax(predicted_classes, axis=1)
loss, acc = model.evaluate(x_test, y_test)
print('\nTesting loss: %.2f, acc: %.2f%%'%(loss, acc))
```
